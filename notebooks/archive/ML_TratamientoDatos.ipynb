# Tratamiento Datos

```python
import pandas as pd
import numpy as np

df = pd.read_excel('Five Lands Stats.xlsx')
```

## Seleccion de columnas

```python
df.columns = df.columns.str.replace('\n', ' ')
df.columns = df.columns.str.replace('  ', '')
df.columns = df.columns.str.lower()
df.drop(columns=['currency',
                  'tuition year',
                  'aid year',
                  'aid year (eur)',
                  'aid year (usd)',
                  'final tuition year',
                  'final tuition year (eur)',
                  'final tuition year (usd)',
                  "% aid",
                  "area of study",
                  "school grade",
                  "last year institution",
                  "type of student",
                  "unnamed: 9",
                  "unnamed: 14"], inplace=True)

df = df.rename(columns={
        "sat (superscore)" : "sat",
        "toefl(max)" : "toefl",
        "profile (4-16)" : "profile",
        "school curriculum" : "bach",
})
```

## Tratamiento de nulos

```python
#area of study
df = df.dropna(subset=['degree'])

#admission
df = df.dropna(subset=['admission'])

#peak
for i in range(len(df["peak"].unique())):
        df["peak"] = df["peak"].replace(df["peak"].unique()[i], i)

#sat
df['sat_required'] = df['sat'].notna().astype(int)
df['sat'] = df['sat'].fillna(0)
```

## Valorizar columnas

```python
#admission
df['admission'] = df['admission'].replace({
    'NO PLAN': 'NO',
    'SELECTED': 'YES'
})
df['admission'] = df['admission'].map({'NO': 0, 'YES': 1})

#profile
df['profile'] = 20 - df['profile']
df.head()
```

## Exportacion

```python
df.to_pickle('ML_Trat.pkl')
```
